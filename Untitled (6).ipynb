```python
# 6)
from PIL import Image,ImageDraw
from IPython.display import display

img1=Image.open("iguane.jpg")
largeur=img1.width
longueur = img1.height
img3=Image.new("RGB",(largeur,longueur),(255,255,255))

for i in range (0,largeur):
    couleur=img1.getpixel((i,longueur))
    R,V,B = couleur
    img3.putpixel((i,0+1),(R,V,B))
    for i in range (i, longueur):
        couleur=img1.getpixel((0,i))
        R,V,B = couleur
        img3.putpixel((0,i),(R,V,B))




display(img1)
display(img3)
```
